### Part A - Text Messages

This code simulates sending text messages every 10 seconds in a WhatsApp group and recording the received network traffic. Different sizes of text messages will be sent, including short, medium, and long messages. The goal is to analyze the network traffic pattern and observe how packet lengths correlate with message sizes. The resulting graph will provide insights into the relationship between message length and network traffic. 

A total of 12 text messages will be sent, consisting of 4 messages of each size category. It's important to note that while analyzing the graph, the differences might not be entirely clear due to additional packets related to traffic.

- short message (10 letters)
- medium message (100 letters)
- long message (500 letters) 

```python
# upload the data
text = pd.read_csv(r"C:\Users\97253\Downloads\Final_text.csv")
```

```python
text
```

We will see all the packets sent without connecting appropriate packets

```python
plt.figure(figsize=(20, 10))  # Adjust the figure size

plt.stem(text['Time'], text['Length'])
plt.title("Text Messages Analysis")
plt.xlabel("Time (seconds)")
plt.ylabel("Length")

plt.grid(True, linestyle='--', alpha=1)  # Add crisscross grid lines

plt.show()
```

In the analysis, a notable pattern emerges: clusters of 3 packets appear in distinct intervals of approximately 10 seconds, forming rounds of three different message sizes.

It's crucial to highlight that each of these clusters comprises precisely three packets. While the presentation displays individual packets, a deeper insight arises when we consider that every group of packets transmitted simultaneously represents a single cohesive message. This realization prompts us to aggregate these packets, allowing us to discern the complete size of each message.

By observing these clusters, we can infer that packets sent within a closely-knit time window share a significant connection. This connection likely stems from their proximity in transmission, implying a deliberate grouping strategy. This strategy aligns with the three distinct sizes of messages, which further underscores the deliberate organization behind the traffic.

In essence, this analysis uncovers a structured approach to message transmission, wherein clusters of packets, each consisting of three distinct sizes, are sent in tandem. This method not only offers a more coherent perspective on network traffic but also sheds light on the underlying communication dynamics within the WhatsApp group.


```python
time_threshold = 0.1
# Group rows based on time difference
grouped = text.groupby((text['Time'].diff() > time_threshold).cumsum())
result = grouped.agg({'Time': 'mean', 'Length': 'sum'})
result['Time'] = result['Time'].astype(int)
grouped.size()
```

```python
plt.figure(figsize=(20, 10))  # Adjust the figure size

plt.bar(result['Time'], result['Length'])
plt.title("Text Messages Analysis")
plt.xlabel("Time (seconds)")
plt.ylabel("Length")

plt.grid(True, linestyle='--', alpha=0.7)  # Add crisscross grid lines

plt.show()
```

Upon examining the graph, a clear trend emerges: Wireshark elegantly organizes the transmitted packets of messages into distinct branches, aligning precisely with our intentional message sizes. These sizes encompass three tiers, which notably correspond to increasing orders of magnitude:

- For a succinct message of 10 letters, the resulting packet length stabilizes around 500 bytes.
- A moderate-sized message containing 100 letters yields packets averaging approximately 800 bytes.
- Lastly, a substantial message encompassing 500 letters leads to packets that stretch to around 1200 bytes.

This striking alignment illustrates how the size of the packets escalates proportionally alongside the message size. The meticulous branching mechanism within Wireshark deftly encapsulates this intricate relationship between message content and packet dimensions.

From these observations, we can confidently delineate the key characteristics underlying the process of sending a text message:

1. **Triplet of Packets:** Each sent text message culminates in the reception of precisely three packets. This consistency across all message sizes indicates a systematic grouping strategy implemented during transmission.

2. **Total Size Range:** The cumulative size of these three packets consistently falls within the range of 500 to 1200 bytes. Notably, in cases of exceptionally lengthy messages, this size might exceed 1200 bytes.

In essence, these findings encapsulate the essence of our analysis. The harmonious correlation between message characteristics and packet behavior enriches our understanding of the communication dynamics within the WhatsApp group.

```python
# upload the data
mix = pd.read_csv(r"C:\Users\97253\Downloads\Final_mix.csv")
```

```python
mix
```

```python
plt.figure(figsize=(20, 10))  # Adjust the figure size

plt.stem(mix['Time'], mix['Length'])
plt.title("Mix Messages Analysis")
plt.xlabel("Time (seconds)")
plt.ylabel("Length")

plt.grid(True, linestyle='--', alpha=1)  # Add crisscross grid lines

plt.show()
```

As you can see we received a wide variety of different sizes for the information, the reason is of course that the information contains many types of messages

```python
time_threshold = 0.09
# Group rows based on time difference
grouped = mix.groupby((mix['Time'].diff() > time_threshold).cumsum())
result = grouped.agg({'Time': 'mean', 'Length': 'sum'})
result['Time'] = result['Time'].astype(int)
grouped.size()
```

```python
result['size'] = grouped.size() 
result['Target'] = None
for index, row in result.iterrows():
    if 500 <= row['Length'] <= 1200 and row['size'] == 3:
        result['Target'][index] = 'Text'
    elif 1200 <= row['Length'] <= 3500 and 3 <= row['size'] <= 9:
        result['Target'][index] = 'File'
    elif 4000 <= row['Length'] <= 6500 and 8 <= row['size'] <= 11:
        result['Target'][index] = 'Video'
    elif 6500 <= row['Length'] <= 8500 and 11 <= row['size'] <= 15:
        result['Target'][index] = 'Image'
    else:
        result['Target'][index] = 'Noise'
```

```python
# Define colors for each target type
target_colors = {
    'Text': 'blue',
    'Image': 'green',
    'File': 'red',
    'Video': 'purple',
    'Noise': 'black'
}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(20,10))

# Iterate through the DataFrame to plot each packet
for idx, row in result.iterrows():
    ax.bar(row['Time'], row['Length'], color=target_colors[row['Target']])

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Length of Packet')
ax.set_title("Mix Messages Analysis")

# Create a custom legend for target types
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=target_colors[target], label=target) for target in target_colors]
ax.legend(fontsize='x-large',loc=9,handles=handles, title='Target Types')
ax.grid(True)
# Show the plot
plt.show()
```

```python

def generate_and_plot_pdf_with_exponential_fit(df, bins=10):

    df_cleaned = df.dropna(subset=['Time_Difference'])
    
    counts, bin_edges = np.histogram(df_cleaned['Time_Difference'], bins=bins, density=True)
    
    # Convert counts to probabilities
    pdf = counts / sum(counts)
    
    plt.figure(figsize=(10, 6))
    
    # Plotting the outline of the histogram using plt.step
    plt.step(bin_edges[:-1], pdf, where='post', color='blue', lw=2, label='PDF')
    
    # Estimate lambda (rate parameter) from the data
    lambd = 1. / np.mean(df['Time_Difference'])
    
    # Compute the bin width from the first two bin edges
    bin_width = bin_edges[1] - bin_edges[0]
    
    # Adjust the scale_factor with respect to bin_width
    scale_factor = max(pdf) / (lambd * np.exp(-lambd * bin_edges[np.argmax(pdf)]))
    
    # Create x values for the fitted exponential curve
    x = np.linspace(0, max(bin_edges), 500)
    
    # Calculate exponential distribution values using the estimated lambda
    y = scale_factor * lambd * np.exp(-lambd * x)
    
    plt.plot(x, y, color='red', lw=2, label='Fitted Exponential Distribution')
    
    plt.xlabel('Inter-Arrival Time')
    plt.ylabel('Density')
    plt.title('PDF of Inter-Arrival Time and Fitted Exponential Distribution')
    plt.legend()
    plt.grid(True)
    
    plt.show()
```
